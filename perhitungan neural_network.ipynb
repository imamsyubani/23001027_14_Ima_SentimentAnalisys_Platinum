```python
import math
import numpy as np
np.set_printoptions(suppress=True)
```

# Initialize value

```python
# Given values
x1 = 0.7
x2 = 0.8
x3 = 0.9

w14 = 0.5
w24 = 0.3
w34 = -1.0
w15 = 0.6
w25 = 1.1
w35 = 0.1

w46 = -1.1
w56 = -0.7

theta4 = 0.2
theta5 = 0.3
theta6 = 0.4
thresold = -1

actual_output = 0
learning_rate = 0.1
```

```python
# Activation function sigmoid
# Sigmoid function
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

# Forward Pass

```python
# Calculate the output for neuron 3 and neuron 4 in the hidden layer
output4 = sigmoid(w14*x1 + w24*x2 + w34*x3 + theta4)
output5 = sigmoid(w15*x1 + w25*x2 + w35*x3 - theta5)

# Calculate the output for neuron 5 in the output layer
output6 = sigmoid(w46*output4 + w56*output5 - theta6)

output4, output5, output6
```

# Backpropagation

```python
# Sigmoid derivative function
def sigmoid_derivative(x):
    return x * (1 - x)
```

```python
# Calculate gradient for neuron 5
gradient_6 = (actual_output - output6) * sigmoid_derivative(output6)
print('Gradient 6:', '{0:.4f}'.format(gradient_6))
```

```python
# Calculate the derivative of the error with respect to the weights
dError_dWeight46 = gradient_6 * output4
dError_dWeight56 = gradient_6 * output5
dError_dtheta6 = gradient_6 * thresold

print('DError:', '{0:.4f}'.format(dError_dtheta6))

# Calculate the weight correction for each weight
delta_weight46 = learning_rate * dError_dWeight46
delta_weight56 = learning_rate * dError_dWeight56
delta_theta6 = learning_rate * dError_dtheta6


print('Delta W46:', '{0:.4f}'.format(delta_weight46))
print('Delta W56:', '{0:.4f}'.format(delta_weight56))
print('Delta Theta6:', '{0:.4f}'.format(delta_theta6))
```

```python
# Update the weights
new_w46 = w46 + delta_weight46
new_w56 = w56 + delta_weight56
new_theta6 = theta6 + delta_theta6

print('New W46:', '{0:.4f}'.format(new_w46))
print('New W56:', '{0:.4f}'.format(new_w56))
print('New Theta6:', '{0:.4f}'.format(new_theta6))
```

```python
# Calculate gradient
gradient_4 = sigmoid_derivative(output4) * gradient_6 * w46
gradient_5 = sigmoid_derivative(output5) * gradient_6 * w56

print('Gradient 4:', '{0:.4f}'.format(gradient_4))
print('Gradient 5:', '{0:.4f}'.format(gradient_5))
```

```python
# Calculate the delta for each weight
delta_weight14 = learning_rate * x1 * gradient_4
delta_weight24 = learning_rate * x2 * gradient_4
delta_weight34 = learning_rate * x3 * gradient_4
delta_weight15 = learning_rate * x1 * gradient_5
delta_weight25 = learning_rate * x2 * gradient_5
delta_weight35 = learning_rate * x3 * gradient_5
delta_theta4 = learning_rate * thresold * gradient_4
delta_theta5 = learning_rate * thresold * gradient_5

print('Delta W14:', '{0:.4f}'.format(delta_weight14))
print('Delta w24:', '{0:.4f}'.format(delta_weight24))
print('Delta w34:', '{0:.4f}'.format(delta_weight34))
print('Delta w15:', '{0:.4f}'.format(delta_weight14))
print('Delta w25:', '{0:.4f}'.format(delta_weight24))
print('Delta w35:', '{0:.4f}'.format(delta_weight35))
print('Delta theta4:', '{0:.4f}'.format(delta_theta4))
print('Delta theta5:', '{0:.4f}'.format(delta_theta5))
```

```python
# Update the weights
new_w14 = w14 + delta_weight14
new_w24 = w24 + delta_weight24
new_w34 = w34 + delta_weight34
new_w15 = w15 + delta_weight15
new_w25 = w25 + delta_weight25
new_w35 = w35 + delta_weight35
new_theta4 = theta4 + delta_theta4
new_theta5 = theta5 + delta_theta5


print('New W14:', '{0:.4f}'.format(new_w14))
print('New W24:', '{0:.4f}'.format(new_w24))
print('New W34:', '{0:.4f}'.format(new_w34))
print('New W15:', '{0:.4f}'.format(new_w15))
print('New W25:', '{0:.4f}'.format(new_w25))
print('New W35:', '{0:.4f}'.format(new_w35))

print('New W46:', '{0:.4f}'.format(new_w46))
print('New W56:', '{0:.4f}'.format(new_w56))

print('New Theta4:', '{0:.4f}'.format(new_theta4))
print('New Theta5:', '{0:.4f}'.format(new_theta5))
print('New Theta6:', '{0:.4f}'.format(new_theta6))
```
